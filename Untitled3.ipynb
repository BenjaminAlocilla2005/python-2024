```python
import numpy as np
import matplotlib.pyplot as plt
```

1. Generar un vector aleatorio que incluya números negativos desde el -10 hasta el 10 positivo.
Este vector debe tener 25 elementos. Deben ser números flotantes.

```python
### Código Aquí
vector = np.random.randint(-10,10,25)
vector
```

2. Transformar el vector en una matriz cuadrada de 5 x 5. Obtener el promedio de la matriz

```python
### Código Aquí
matriz = vector.reshape(5,5)
matriz
```

3. Generar un mapa de calor a partir de la matriz generada anteriormente

```python
### Código Aquí
fig, grafico = plt.subplots()

cgrafico = grafico.imshow(matriz, cmap='hot', interpolation='nearest')

fig.colorbar(cgrafico)

grafico.set_title("Mapa de Calor de una Matriz Aleatoria")
grafico.set_xlabel("Columnas")
grafico.set_ylabel("Filas")

plt.show
```
